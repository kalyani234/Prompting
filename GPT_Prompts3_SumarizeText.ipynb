# Sumarizing review

````python
from google.colab import userdata
from openai import OpenAI

client = OpenAI(api_key=userdata.get("open_api"))

def get_completion(prompt, model="gpt-4o-mini"):
    response = client.chat.completions.create(
        model=model,
        messages=[{"role": "user", "content": prompt}],
        temperature=0
    )
    return response.choices[0].message.content
````

## The price and perceived value

##

````python
prompt = f"""
Your task is to generate a short summary of a product
review from an ecommerce site to give feedback to the
pricing deparmtment, responsible for determining the
price of the product.

Summarize the review below, delimited by triple
backticks, in at most 30 words, and focusing on any aspects
that are relevant to the price and perceived value.

Review: {prod_review}
"""

response = get_completion(prompt)
print(response)
````

## Summary for multiple product reviews\

````python
for i in range(len(reviews)):
    prompt = f"""
    Your task is to generate a short summary of a product
    review from an ecommerce site.

    Summarize the review below, delimited by triple
    backticks in at most 20 words.

    Review: ```{reviews[i]}```
    """

    response = get_completion(prompt)
    print(i, response, "\n")
````

## Sumarize in bullet points

````python
for i in range(len(reviews)):
    prompt = f"""
    Your task is to generate a short summary of a product
    review from an ecommerce site describing the highlights of products.

    Summarize the review below, delimited by triple
    backticks in 2 bullet points each in 10 words.

    Review: ```{reviews[i]}```
    """

    response = get_completion(prompt)
    print(i, response, "\n")
````
